# 05 Monitoring

Compare training reference data against current inference data and inspect prediction distribution shift.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.append(str(PROJECT_ROOT))

from src.customer_drift import run_monitoring

drift_report = run_monitoring(
    reference_path=PROJECT_ROOT / "data" / "processed" / "train.csv",
    current_path=PROJECT_ROOT / "data" / "inference" / "weekly_scoring_batch.csv",
    predictions_path=PROJECT_ROOT / "outputs" / "predictions" / "customer_risk_predictions.csv",
    output_path=PROJECT_ROOT / "outputs" / "monitoring" / "drift_report.csv",
)
drift_report.head(20)
```

```python
drift_report.assign(psi_numeric=lambda x: __import__("pandas").to_numeric(x["psi_score"], errors="coerce"))\
    .dropna(subset=["psi_numeric"])\
    .sort_values("psi_numeric", ascending=False)\
    .head(10)[["metric_type", "feature", "category", "psi_score", "drift_severity"]]
```

Data drift tracks input feature changes. Model quality drift requires actual labels for the scored period. Concept drift means the business relationship between customer behavior and risk has changed.
