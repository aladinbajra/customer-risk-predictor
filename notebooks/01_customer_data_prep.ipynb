# 01 Data Preparation

Generate or load the customer risk snapshot, create the default inference sample, and generate EDA artifacts. This notebook is Databricks-friendly and calls reusable code from `src/`.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.append(str(PROJECT_ROOT))

import pandas as pd
from src.customer_data_factory import generate_customer_risk_data, write_dataset, generate_inference_sample
from src.risk_metrics import create_eda_artifacts
from src.customer_features import TARGET_COLUMN

DATA_PATH = PROJECT_ROOT / "data" / "raw" / "customer_risk_snapshot.csv"
INFERENCE_PATH = PROJECT_ROOT / "data" / "inference" / "weekly_scoring_batch.csv"

if not DATA_PATH.exists():
    df_generated = generate_customer_risk_data(rows=15000, seed=42)
    write_dataset(df_generated, DATA_PATH)
    write_dataset(generate_inference_sample(df_generated), INFERENCE_PATH)

df = pd.read_csv(DATA_PATH)
df.shape
```
