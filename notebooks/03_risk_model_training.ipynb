# 03 Model Training with MLflow

Train and compare Logistic Regression, Random Forest, and HistGradientBoosting with local MLflow tracking.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.append(str(PROJECT_ROOT))

import pandas as pd

from src.risk_metrics import create_eda_artifacts, predict_probabilities, save_confusion_matrix_figure, save_roc_curve_figure, split_features_target
from src.train_risk_model import split_dataset, save_splits, train_and_compare, save_model_artifacts

df = pd.read_csv(PROJECT_ROOT / "data" / "raw" / "customer_risk_snapshot.csv")
create_eda_artifacts(df, PROJECT_ROOT / "outputs" / "figures")
train_df, validation_df, test_df = split_dataset(df)
save_splits(train_df, validation_df, test_df)

best_model_name, best_pipeline, comparison_df, final_metrics, run_id = train_and_compare(
    train_df=train_df,
    validation_df=validation_df,
    test_df=test_df,
    experiment_name="customer_risk_predictor",
    tracking_uri="file:./mlruns",
)

X_test, y_test = split_features_target(test_df)
probabilities = predict_probabilities(best_pipeline, X_test)
save_confusion_matrix_figure(y_test, probabilities, PROJECT_ROOT / "outputs" / "figures" / "champion_confusion_matrix.png")
save_roc_curve_figure(y_test, probabilities, PROJECT_ROOT / "outputs" / "figures" / "champion_roc_curve.png")
save_model_artifacts(best_pipeline, final_metrics, comparison_df)

best_model_name, final_metrics
```
